#### filling the missing values with mode

```python
df.A1.fillna(df.A1.mode()[0],inplace=True)
df.A1.isnull().sum()
```

#### i will perform one-hot-encoding

```python
A1_dummy = pd.get_dummies(df.A1)
print(df.shape)
df.drop('A1',inplace=True,axis=1)
print(df.shape)
df = pd.concat([df,A1_dummy])
df.shape
```
